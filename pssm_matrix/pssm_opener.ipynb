```python
import pickle
import os

def open_and_read_pkl(file_path):
    """Opens a .pkl file and reads its contents.

    Args:
        file_path: The path to the .pkl file.

    Returns:
        The contents of the .pkl file.
    """
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
    return data

# Get a list of all .pkl files in the current directory
directory = '/home/bibhuprasad/Documents/PPI prediction model/PPI-predictor-/pssm_matrix'
pkl_files = [f for f in os.listdir(directory) if f.endswith('.pkl')]

# Iterate through the .pkl files and print their contents
for file_name in pkl_files:
    file_path = os.path.join(directory, file_name)
    data = open_and_read_pkl(file_path)
    print(f"Contents of {file_name}:")
    print(data)
    print("-" * 20)  # Separator between files
```
